````julia
using TestParticle
using CSV
using DataFrames
using Plots
theme(:vibrant)
default(frame_style=:box)
````

````julia
df = CSV.read("SN_m_tot_V2.0.csv", delim=";", header=1,  DataFrame)
t = df[:, :FracOfYear]
SN = df[:, :SN];
````

````julia
plot(t, SN, label="")
xlabel!("Jahr")
ylabel!("Sonnenfleckenrelativzahl")
````

````julia
using FFTW
using Statistics
````

````julia
S = fft(SN .- mean(SN));
````

````julia
popfirst!(S)
plot(abs.(S))
````

````julia
n = length(S)
up = Int(floor(n/2))
power = abs.(S[1:up, :]).^2
samplinginterval = 1.0 / 12.0
nyquist = 1 / (2 * samplinginterval)
freq = [1:up...] ./ (n / 2) * nyquist;
````

````julia
plot(freq, power)
xlabel!("Cycles per year")
````

````julia
period = 1 ./ freq
i = argmax(power)

plot(period, power, xscale=:log10, label="")
scatter!([period[i]], [power[i]], m=:o, label="")
xlabel!("Years")
title!("Prominent cycle: $(round(period[i], digits=2)) years")
````
